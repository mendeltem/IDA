```python
def one_hot(data, colname):
    temp = data.loc[:,colname].unique()
    dictionary = dict((name,i) for i,name in enumerate(temp) if "?" != name and " ?" != name)
    data[colname] = data[colname].map(dictionary)
    return data,dictionary
```
